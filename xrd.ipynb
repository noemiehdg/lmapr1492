Tâche 7 : xrd, pics d'un diffractograme
===============================

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
#Téléchargement du fichier CIF de notre matériau
with MPRester("k7Oj07sVAFLX1l2ahe") as m:
    structure = m.get_structure_by_material_id("mp-15790")
```

L'objectif de cette tâche est d'analyser les 3 premiers pics du difractogramme correspondant (en indiquant les indices hkl associés) pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.

Pour nous aider, nous allons utiliser pymatgen, matérials project et la théorie du chapitre 3 sur la radiocristallographie du cours de physique des matériaux.



Affichage des pics 
-------------------------------

```python
#création de la figure de diffraction 
xrd_CuKa= XRDCalculator('CuKa1')

#on peut vérifier que ce soit la bonne longueur d'onde
xrd_CuKa.wavelength
```

```python
#plot du diagramme (pour 0 à 30 degré, les quatres premiers pics sont inclus dedans)
xrd_CuKa.show_plot(structure,fontsize=25,two_theta_range=(0,30))
```

On compare ce diagramme à celui de Materials Project



Pour comparer ces deux graphiques, il est important de garder en tête que le matérials project se base sur la maille conventionelle et que les calculs effectués ici se font dans la maille primitive (les indices hkl's sont donc différents). Cependant, on remarque que les angles d'incidences et l'intensité de ceux-ci semblent correspondre, nous allons maintenant les étudier plus en détail.

Analyse des trois pics 
-------------------------------

```python
pattern= xrd_CuKa.get_pattern(structure)
hkls=pattern.hkls
twotheta=pattern.x
intensity=pattern.y
```

### Premier pic

```python
print("Angle d'incidence : " + str(twotheta[0]) + "°")
print("Intensité : " + str(intensity[0]))
print("Indices hkl associés et sa multiplicité " + str(hkls[0]))
```

NB. La multiplicité 2 nous indique qu'il existe deux plans d'indices de Miller, équivalents par symétrie.
Pour ce cas-ci, le plan (1, 1, 1) équivaut au plan (-1, -1, -1)

```python
print("Angle d'incidence : " + str(twotheta[1]) + "°")
print("Intensité : " + str(intensity[1]))
print("Indices hkl associés et sa multiplicité " + str(hkls[1]))
```
